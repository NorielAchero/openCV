**openCV Functions**

```python
# Functions

# cv2.imread() - Used to input an image to work with / cv2.imread(file path, flag)
# cv2.imshow() - Displays the image in the working environment / cv2.imshow(windowname, image to be displayed)
# cv2.imwrite() - Write any changes to an actual image file / cv2.imwrite(filename, image to be saved)
```

**Reading an Image and printing the dimension:**

```python
image = cv2.imread('/content/sample_data/sample.jpg')

x, y = image.shape[:2]

print("Height = {}, Width = {}".format(x,y))
```

**Extracting RGB Values of the pixels**

```python
(R, G, B) = image[250, 250]

print("R = {}, G = {}, B = {}".format(R, G, B))

B = image[250, 250, 0]
print("Just the value of B = {}".format(B))
```

Resizing the image to any dimension:

```python
size = cv2.resize(image, (500, 500))
cv2.imwrite("resized.jpg", size)
```

```python
resized = cv2.imread('/content/resized.jpg')

x, y = resized.shape[:2]

print("Height = {}, Width = {}".format(x,y))
```

**Placing a box around any pixel**

```python
rectangle = cv2.rectangle(image, (200, 900), (400, 400), (255, 0, 0), 2)

cv2.imwrite("rectangled.jpg", image)
```

**Adding text to the screen**

```python
text_input = cv2.putText(image, 'KitKat', (300, 300), cv2.FONT_HERSHEY_SIMPLEX, 4, (255, 0, 0), 2)

cv2.imwrite("text.jpg", image)
```
